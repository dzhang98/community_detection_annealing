```python
import os, sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
os.chdir("../result_files")

dataset = "sp100" # Values: asset_class, sectors, ftse

SA_res_df = pd.read_csv("modularity_SA_%s_res.csv" % (dataset))
louvain_res_df = pd.read_csv("modularity_louvain_%s_res.csv" % (dataset))
igraph_res_df = pd.read_csv("modularity_igraph_%s_res.csv" % (dataset))
```

```python
SA_res_df = SA_res_df.drop(axis=1, columns=['Unnamed: 0'])
louvain_res_df = louvain_res_df.drop(axis=1, columns=['Unnamed: 0'])
igraph_res_df = igraph_res_df.drop(axis=1, columns=['Unnamed: 0'])
```

```python
plt.figure(figsize=(20, 10))
plt.title('Modularity vs Date for Dataset %s' % (dataset))
plt.ylabel('Modularity')
plt.xlabel('Date')

plt.plot(SA_res_df["date"], SA_res_df['modularity'],'g--', label='DWave SA Algorithm')
plt.plot(louvain_res_df["date"], louvain_res_df['modularity'], 'r--', label='Louvain Algorithm')
plt.plot(igraph_res_df["date"], igraph_res_df['modularity'], 'b--', label='iGraph SA Algorithm')

plt.legend(loc=0, prop={'size': 10})
plt.xticks(ticks=SA_res_df['date'][::4], 
              labels=[x[0:7] for x in SA_res_df['date']])
```

```python
plt.figure(figsize=(20, 10))
plt.title('Computation Time vs Date for Dataset %s' % (dataset))
plt.ylabel('Computation Time')
plt.xlabel('Date')

plt.plot(SA_res_df["date"], SA_res_df['computation_time'],'g--', label='DWave SA Algorithm')
plt.plot(louvain_res_df["date"], louvain_res_df['computation_time'], 'r--', label='Louvain Algorithm')
plt.plot(igraph_res_df["date"], igraph_res_df['computation_time'], 'b--', label='iGraph SA Algorithm')

plt.legend(loc=0, prop={'size': 10})
plt.xticks(ticks=SA_res_df['date'][::4], 
              labels=[x[0:7] for x in SA_res_df['date']])
```
